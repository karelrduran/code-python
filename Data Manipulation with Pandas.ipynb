### Data Manipulation with Pandas

```python
# imports
import pandas as pd
import numpy as np
```

```python
# Read a CSV file
df = pd.read_csv('data/taxis.csv')
df
```

```python
# Display dataframe's columns
df.columns
```

```python
# Filter where clause
cash_payments = df[df['payment'] == 'cash']
cash_payments
```

```python
# Just select the columns: pickup_zone, dropoff_zone
cash_payments = cash_payments[['pickup_zone', 'dropoff_zone']]
cash_payments
```

```python
# Put the filtered data into a new csv file
cash_payments.to_csv('data/filtered_cash_payments_with_pickup_dropoff.csv', index=False)
```
